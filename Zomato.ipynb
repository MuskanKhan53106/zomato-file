#**2. Data cleaning**

#####**2.1 checking the missing values**

```python
data.isnull().sum()
```

â–¶there are so many null values .we can clearly see that in  'rate', 'phone', 'location', 'rest_type', 'dish_liked', 'cuisines' and 'approx_cost(for two people)' these columns have the missing values .firstly we have have to handle the missing values.

```python
data.isnull().sum()
```

##### **2.3 Removing the unnecessary columns**

```python
data.drop(['url','phone'],axis=1,inplace=True)
```

```python
data.duplicated().sum()
```

```python
data.drop_duplicates(inplace=True)
```

```python
data.duplicated().sum()
```

â–¶ Now there is no duplicates values

###### **2.5 Renaming the columns appropriately**

```python
data=data.rename(columns={'approx_cost(for two people)':'cost',
                         'listed_in(type)':'type', 'listed_in(city)': 'city'})
```

```python
data.head()
```

â–¶sucessfully rename the columns

###### **2.6 clean the "cost" column**

```python
data["cost"].unique()
```

â–¶ here we can see that data point is string type and some value like 5,000 6,000 have comma(,).we have to remove that comma from tha values and we have convert them into numeric type.

```python
data["cost"]=data["cost"].apply(lambda x:x.replace(",",""))
data["cost"]=data["cost"].astype(float)

data["cost"].unique()
```

â–¶ now sucessfully we converted values into numeric type

##### **2.7 handling the rate columns**

```python
data['rate'].unique()
```

â–¶here in rating column also string type so we have to convert them into numeric type.first we have ti remove '/5' from values and there is 'NEW' value which make no sense here. so we have to remove that value.

```python
data=data.loc[data.rate != 'NEW']
data['rate'].unique()
```

```python
data['rate']=data['rate'].apply(lambda x:x.replace('/5',''))
data['rate'].unique()
```

â–¶ now our data is clean we can perform visualization.

# **4. Data Preparing**

```python
data.head()
```

##### **4.1 Convert the online categorical into numeric format**

```python
data.online_order[data.online_order == 'Yes'] = 1
data.online_order[data.online_order == 'No'] = 0

```

```python
data.head(2)
```

```python
data.online_order.value_counts()
```

##### **4.2 Label Encoding**

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
data.location=le.fit_transform(data.location)
data.rest_type=le.fit_transform(data.rest_type)
data.cuisines=le.fit_transform(data.cuisines)
data.menu_item=le.fit_transform(data.menu_item)
data.book_table=le.fit_transform(data.book_table)
```
